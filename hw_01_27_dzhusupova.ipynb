–£–®–≠ (–£—Ä—é–ø–∏–Ω—Å–∫–∞—è –®–∫–æ–ª–∞ –≠–∫–æ–Ω–æ–º–∏–∫–∏) –Ω—ã–Ω–µ –ø—Ä–µ—Å—Ç–∏–∂–Ω–∞ –∏ –∑–∞–±–∏—Ä–∞–µ—Ç —Å–µ–±–µ –≤—Å–µ—Ö —à–∫–æ–ª—å–Ω–∏–∫–æ–≤ –Ω–∞–±—Ä–∞–≤—à–∏—Ö –±–æ–ª–µ–µ $n$ –±–∞–ª–ª–æ–≤ –≤ —Å—É–º–º–µ –ø–æ —Ä—É—Å—Å–∫–æ–º—É –∏ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ. 

–î–ª—è –∫–∞–∂–¥–æ–≥–æ $n$ –æ—Ç $80$ –¥–æ $160$ —Å —à–∞–≥–æ–º $5$ —Å–ª—É—á–∞–π–Ω–æ —Å–æ–∑–¥–∞–π—Ç–µ $10^4$ —à–∫–æ–ª—å–Ω–∏–∫–æ–≤ –∏ –ø–æ–¥–µ–ª–∏—Ç–µ –∏—Ö –Ω–∞ –ø—Ä–æ—à–µ–¥—à–∏—Ö –∏ –Ω–µ –ø—Ä–æ—à–µ–¥—à–∏—Ö –≤ –£–®–≠.


__–∞) [5 –±–∞–ª–ª–æ–≤]__ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —á–∏—Å–ª–∞ –ø—Ä–æ—à–µ–¥—à–∏—Ö –≤ –£–®–≠ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç $n$. 

__–±) [5 –±–∞–ª–ª–æ–≤]__ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –º–µ–∂–¥—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –ø–æ —Ä—É—Å—Å–∫–æ–º—É –∏ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç $n$ —Å—Ä–µ–¥–∏ –ø—Ä–æ—à–µ–¥—à–∏—Ö –≤ –£–®–≠.

__–≤) [5 –±–∞–ª–ª–æ–≤]__ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –º–µ–∂–¥—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –ø–æ —Ä—É—Å—Å–∫–æ–º—É –∏ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç $n$ —Å—Ä–µ–¥–∏ –Ω–µ –ø—Ä–æ—à–µ–¥—à–∏—Ö –≤ –£–®–≠.

–í —á—ë–º –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –ø–∞—Ä–∞–¥–æ–∫—Å? –ü—Ä–æ–∫–æ–º–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–∏–≤—à–∏–µ—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

```python
scholars = [(sps.norm(loc=60, scale=10).rvs(size=10000), sps.norm(loc=60, scale=10).rvs(size=10000)) for i in range(80, 165, 5)]
```

```python
scholars_success = []
for idx, i in enumerate(scholars):
  elts = []
  for j in range(10000):
    if i[0][j] + i[1][j] >= 80 + idx * 5:
      elts.append((i[0][j],i[1][j]))
  scholars_success.append(elts)
```

```python
scholars_fail = []
for idx, i in enumerate(scholars):
  elts = []
  for j in range(10000):
    if i[0][j] + i[1][j] < 80 + idx * 5:
      elts.append((i[0][j],i[1][j]))
  scholars_fail.append(elts)
```

```python
plt.plot(list(range(80, 165, 5)), [len(i) for i in scholars_success], 'gs')
plt.xlabel('n')
plt.ylabel('scholars number')
plt.show()
```

```python
scholars_success_df = pd.DataFrame(data = {'math': [i[0] for i in scholars_success[0]],
                                           'lang': [i[1] for i in scholars_success[0]],
                                           'n': 80})
for i in range(1, 17):
  scholars_success_df = pd.concat([scholars_success_df, 
                                   pd.DataFrame(data = {'math': [j[0] for j in scholars_success[i]],
                                                        'lang': [j[1] for j in scholars_success[i]],
                                                        'n': 80 + i * 5
                                                        })])
```

```python
scholars_success_df.sample(5)
```

```python
sns.scatterplot(data=scholars_success_df, x='math', y='lang', hue='n', palette="deep")
```

```python
scholars_fail_df = pd.DataFrame(data = {'math': [i[0] for i in scholars_fail[0]],
                                        'lang': [i[1] for i in scholars_fail[0]],
                                        'n': 80})
for i in range(1, 17):
  scholars_fail_df = pd.concat([scholars_fail_df, 
                                pd.DataFrame(data = {'math': [j[0] for j in scholars_fail[i]],
                                                     'lang': [j[1] for j in scholars_fail[i]],
                                                     'n': 80 + i * 5
                                                     })])
```

```python
scholars_fail_df.sample(5)
```

```python
sns.scatterplot(data=scholars_fail_df, x='math', y='lang', hue='n', palette="deep")
```

### –ó–∞–¥–∞—á–∞ 2 (–ü–∞—Ä–∞–¥–æ–∫—Å –∏–Ω—Å–ø–µ–∫—Ç–æ—Ä–∞)

–ê–≤—Ç–æ–±—É—Å—ã –æ—Ç—Ö–æ–¥—è—Ç –æ—Ç –∞–≤—Ç–æ—Å—Ç–∞–Ω—Ü–∏–∏ —Å 8:00 –¥–æ 20:00. –ü–µ—Ä–≤—ã–π –∞–≤—Ç–æ–±—É—Å –æ—Ç—Ö–æ–¥–∏—Ç —Ä–æ–≤–Ω–æ –≤ 8:00. –ó–∞—Ç–µ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –º–µ–∂–¥—É –∞–≤—Ç–æ–±—É—Å–∞–º–∏ —Å–ª—É—á–∞–π–Ω—ã, –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã –∏ —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–Ω–æ —Ä–∞–≤–Ω—ã –ª–∏–±–æ 5-–∏, –ª–∏–±–æ 10-–∏ –º–∏–Ω—É—Ç–∞–º. –ë—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –∑–∞ –º–∏–Ω—É—Ç—É –Ω–∞ –∞–≤—Ç–æ—Å—Ç–∞–Ω—Ü–∏—é –ø—Ä–∏—Ö–æ–¥–∏—Ç —Ä–æ–≤–Ω–æ –æ–¥–∏–Ω –ø–∞—Å—Å–∞–∂–∏—Ä, –∏ –≤—Å–µ –ø–∞—Å—Å–∞–∂–∏—Ä—ã –µ–¥—É—Ç –±–ª–∏–∂–∞–π—à–∏–º –∞–≤—Ç–æ–±—É—Å–æ–º.

–ü—Ä–æ–≤–µ–¥–∏—Ç–µ $10^4$ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤ –∏ —Å –∏—Ö –ø–æ–º–æ—â—å—é:

__–∞) [5 –±–∞–ª–ª–æ–≤]__ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞–≤—Ç–æ–±—É—Å–æ–≤, –æ—Ç–æ—à–µ–¥—à–∏—Ö –æ—Ç –∞–≤—Ç–æ—Å—Ç–∞–Ω—Ü–∏–∏ –∑–∞ —Å—É—Ç–∫–∏. –ü–æ—Ö–æ–∂–µ –ª–∏ –≤–∏–∑—É–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ?

__–±) [5 –±–∞–ª–ª–æ–≤]__ –ò–Ω—Å–ø–µ–∫—Ç–æ—Ä –¢–∏–º–æ–Ω –≤—ã–±–∏—Ä–∞–µ—Ç —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–Ω–æ –æ–¥–∏–Ω –∏–∑ –≤—Å–µ—Ö –∞–≤—Ç–æ–±—É—Å–æ–≤ –æ—Ç–æ—à–µ–¥—à–∏—Ö –æ—Ç –∞–≤—Ç–æ—Å—Ç–∞–Ω—Ü–∏–∏ –∑–∞ —Å—É—Ç–∫–∏. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —á–∏—Å–ª–∞ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤ –Ω–∞ —ç—Ç–æ–º –∞–≤—Ç–æ–±—É—Å–µ. –û—Ü–µ–Ω–∏—Ç–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é —ç—Ç–æ–≥–æ —á–∏—Å–ª–∞.

__–≤) [5 –±–∞–ª–ª–æ–≤]__ –ò–Ω—Å–ø–µ–∫—Ç–æ—Ä –ü—É–º–±–∞ –ø—Ä–∏—Ö–æ–¥–∏—Ç –Ω–∞ –∞–≤—Ç–æ—Å—Ç–∞–Ω—Ü–∏—é –≤ —Å–ª—É—á–∞–π–Ω—ã–π –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏, —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω—ã–π –æ—Ç 8:00 –¥–æ 20:00 –∏ —Å–∞–¥–∏—Ç—Å—è –≤ –ø–µ—Ä–≤—ã–π –ø—Ä–∏—à–µ–¥—à–∏–π –∞–≤—Ç–æ–±—É—Å. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —á–∏—Å–ª–∞ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤ –Ω–∞ —ç—Ç–æ–º –∞–≤—Ç–æ–±—É—Å–µ. –û—Ü–µ–Ω–∏—Ç–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é —ç—Ç–æ–≥–æ —á–∏—Å–ª–∞.

__–≥) [5 –±–∞–ª–ª–æ–≤]__ –ö–∞–∫ –∏–∑–º–µ–Ω—è—Ç—Å—è –æ—Ç–≤–µ—Ç—ã –Ω–∞ —ç—Ç–∏ –≤–æ–ø—Ä–æ—Å—ã, –µ—Å–ª–∏ –≤—Ä–µ–º—è –º–µ–∂–¥—É –∞–≤—Ç–æ–±—É—Å–∞–º–∏ –±—É–¥–µ—Ç —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ —Å–æ —Å—Ä–µ–¥–Ω–∏–º –≤ 10 –º–∏–Ω—É—Ç?

__–ü–æ—è—Å–Ω–µ–Ω–∏—è:__ –ø–µ—Ä–≤—ã–π –∞–≤—Ç–æ–±—É—Å –º–æ–∂–Ω–æ –Ω–µ –≤–∫–ª—é—á–∞—Ç—å –≤ –≤—ã–±–æ—Ä–∫—É, –ü—É–º–±—É –∑–∞ –ø–∞—Å—Å–∞–∂–∏—Ä–∞ –º–æ–∂–Ω–æ –Ω–µ —Å—á–∏—Ç–∞—Ç—å. 

–í —á—ë–º –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –ø–∞—Ä–∞–¥–æ–∫—Å? –ü—Ä–æ–∫–æ–º–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–∏–≤—à–∏–µ—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

```python
intervals = []
for i in range(10000):
  mins = 0
  elts = []
  while mins <= 720:
    rnd = sps.bernoulli.rvs(0.5, size=1)
    elts.append(rnd[0])
    mins += 5 + 5 * rnd[0]
  intervals.append(elts)
```

```python
bus_df = pd.DataFrame(data = {'interval': intervals[0],
                              'experiment': 1
                              })
for i in range(1, 10000):
  bus_df = pd.concat([bus_df, 
                      pd.DataFrame(data = {'interval': intervals[i],
                                           'experiment': i + 1
                                           })])
```

```python
bus_df.sample(5)
```

```python
bus_df.value_counts().hist()
```

### üêà –∞) –í–∏–∑—É–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Ö–æ–∂–µ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ–∂–∏–¥–∞–Ω–∏–µ–º –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ 45-55.

```python
rnd_bus = [sps.randint.rvs(1, len(i), size=1)[0] for i in intervals]
```

```python
# –±) 
vals = [i[rnd_bus[idx]] * 5 + 5 for idx,i in enumerate(intervals)]
plt.xlabel('# of passengers', fontsize=10) # –ø–æ–¥–ø–∏—Å–∏ –æ—Å–µ–π
plt.ylabel('# of simuls', fontsize=10)
plt.hist(vals);
```

```python
# –±) estimation of expected value
x_bar = np.mean(vals)
print('estimation of expected value =',x_bar)
```

```python
# –±) estimation of dispersion
arr_5 = list(filter(lambda x: x == 5, vals))
arr_10 = list(filter(lambda x: x == 10, vals))
disp_hat = len(arr_5) / len(vals) * len(arr_10) / len(vals) 
print('estimation of dispersion =', disp_hat)
```

```python
# –≤)
uniform_rv = sps.uniform(0, 720)
n_obs = 10**4
y = uniform_rv.rvs(n_obs)
uniform_vals = []
for idx, i in enumerate(intervals):
  tm = 0
  for j in i:
    tm += 5 * (j + 1)
    if y[idx] <= tm:
      uniform_vals.append(tm)
      break
plt.hist(uniform_vals, bins=20);
```

```python
# –≤) estimation of expected value
x_bar = np.mean(uniform_vals)
print('estimation of expected value =',x_bar)
```

```python
# –≤) estimation of dispersion
val = 0
for i in uniform_vals:
  val += (i - x_bar)**2
disp_hat = np.sqrt(1 / (len(uniform_vals) - 1) * val)
print('estimation of dispersion =', disp_hat)
```

```python
# –≥) - –∞
intervals_exp = []
for i in range(10000):
  mins = 0
  elts = []
  while mins <= 720:
    rnd = round(sps.expon.rvs(scale=10, size=1)[0], 2)
    elts.append(rnd)
    mins += rnd
  intervals_exp.append(elts)
```

```python
# –≥) - –∞
bus_exp_df = pd.DataFrame(data = {'interval': intervals_exp[0],
                              'experiment': 1
                              })
for i in range(1, 10000):
  bus_exp_df = pd.concat([bus_exp_df, 
                      pd.DataFrame(data = {'interval': intervals_exp[i],
                                           'experiment': i + 1
                                           })])
```

```python
# –≥) - –∞
bus_exp_df.sample(5)
```

```python
# –≥) - –∞
sns.histplot(data=bus_exp_df, x="interval");
```

```python
# –≥) - –±
rnd_bus_exp = [sps.randint.rvs(1, len(i), size=1)[0] for i in intervals_exp]
```

```python
# –≥) - –± 
import math
vals_exp = [math.floor(i[rnd_bus_exp[idx]]) for idx,i in enumerate(intervals_exp)]
plt.xlabel('# of passengers', fontsize=10) # –ø–æ–¥–ø–∏—Å–∏ –æ—Å–µ–π
plt.ylabel('# of simuls', fontsize=10)
plt.hist(vals_exp, bins=15);
```

```python
# –≥) - –±
x_bar = np.mean(vals_exp)
print('estimation of expected value =', x_bar)
```

```python
# –≥) - –±
val = 0
for i in vals_exp:
  val += (i - x_bar)**2
disp_hat = np.sqrt(1 / (len(vals_exp) - 1) * val)
print('estimation of dispersion =', disp_hat)
```

```python
# –≥) - –≤
uniform_rv = sps.uniform(0, 720)
n_obs = 10**4
y = uniform_rv.rvs(n_obs)
uniform_vals = []
for idx, i in enumerate(intervals_exp):
  tm = 0
  for j in i:
    tm += j
    if y[idx] <= tm:
      uniform_vals.append(tm)
      break
plt.hist(uniform_vals, bins=50);
```

```python
# –≥) - –≤
x_bar = np.mean(uniform_vals)
print('estimation of expected value =', x_bar)
```

```python
# –≥) - –≤
val = 0
for i in uniform_vals:
  val += (i - x_bar)**2
disp_hat = np.sqrt(1 / (len(uniform_vals) - 1) * val)
print('estimation of dispersion =', disp_hat)
```

### –ó–∞–¥–∞—á–∞ 3 (—á—Ç–æ-—Ç–æ –ø–æ—Ö–æ–∂–µ–µ –Ω–∞ –¶–ü–¢, –Ω–æ –≤ –æ–±—Ä–∞—Ç–Ω—É—é —Å—Ç–æ—Ä–æ–Ω—É)

–ê–≤—Ç–æ—Ä—É –¥–æ–º–∞—à–∫–∏ –≤–∫–∏–Ω—É–ª–∏ —ç—Ç—É –∑–∞–¥–∞—á—É –≤ –∫–∞—á–µ—Å—Ç–≤–µ –≤—Å—Ç—É–ø–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏—Å–ø—ã—Ç–∞–Ω–∏—è –Ω–∞ —Å—Ç–∞–∂–∏—Ä–æ–≤–∫—É. –í –ª–∏—Å—Ç–æ—á–∫–µ –±—ã–ª–æ –æ–∫–æ–ª–æ –¥–µ—Å—è—Ç–∏ –ø–æ—Ö–æ–∂–∏—Ö –∑–∞–¥–∞—á. –†–µ—à–∞—Ç—å –Ω–∞–¥–æ –±—ã–ª–æ —Ä—É–∫–∞–º–∏ –±–µ–∑ –∫–æ–º–ø—å—é—Ç–µ—Ä–∞. –ù–æ –º—ã –ø–æ–ø—Ä–æ–±—É–µ–º —Å. 

–ü—É—Å—Ç—å $X \sim N(0, \sigma^2)$. –ü—É—Å—Ç—å $\Phi(x)$ ‚Äî —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è $N(0,1)$. –ù—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã $\Phi(X)$. 

__–∞) [2.5 –±–∞–ª–ª–∞]__ –û—Ü–µ–Ω–∏—Ç–µ $\mathbb{E}(\Phi(X))$ —Å –ø–æ–º–æ—â—å—é —Å–∏–º—É–ª—è—Ü–∏–π

__–±) [2.5 –±–∞–ª–ª–∞]__ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã $\Phi(X)$ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É. –í–æ–∑—å–º–∏—Ç–µ $\sigma = 1$, –ø–æ—Å–ª–µ $\sigma = 3$, –ø–æ—Å–ª–µ $\sigma = 100$. –ö–∞–∫ –≤–µ–¥—ë—Ç —Å–µ–±—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏ —Ä–æ—Å—Ç–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏? –ö–∞–∫ –¥—É–º–∞–µ—Ç–µ, –∫ –∫–∞–∫–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω–µ —Å–æ–π–¥—ë—Ç—Å—è —ç—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ?

```python
grid_sigma = np.linspace(0.1, 5, 100)
x_data = [sps.norm(loc=0, scale=i).rvs(size=10000) for i in grid_sigma]
```

```python
distribution = sps.norm(loc=0, scale=1)
for i in range(len(grid_sigma)):
  sns.set(style='darkgrid')  # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å—Ç–∏–ª—å
  grid = sorted(x_data[i])
  plt.plot(grid, distribution.cdf(grid), label='$F(x)$') # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ª–∏–Ω–∏–∏ F(x)
plt.xlabel('$x$', fontsize=10) # –ø–æ–¥–ø–∏—Å–∏ –æ—Å–µ–π
plt.ylabel('$F(x)$', fontsize=10)
```

### ü¶ä a) –°–æ–≥–ª–∞—Å–Ω–æ –≤–∏–∑—É–∞–ª—å–Ω–æ–º—É –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—é —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Å–∏–º—É–ª—è—Ü–∏–π –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ 10000 —Å —Ä–∞–∑–ª–∏—á–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è –∏ —Ä–∞–≤–Ω–æ –Ω—É–ª—é.

```python
grid_sigma = [1, 3, 100]
x_data = [sps.norm(loc=0, scale=i).rvs(size=10000) for i in grid_sigma]
for i in range(len(grid_sigma)):
  sns.set(style='darkgrid')  # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å—Ç–∏–ª—å
  grid = sorted(x_data[i])
  plt.plot(grid, distribution.cdf(grid), label='$F(x)$') # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ª–∏–Ω–∏–∏ F(x)
plt.xlabel('$x$', fontsize=10) # –ø–æ–¥–ø–∏—Å–∏ –æ—Å–µ–π
plt.ylabel('$F(x)$', fontsize=10)
```

### üê© –±) –ü–æ–¥—Å—Ç–∞–≤–ª—è—è –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –≤ —Å–µ—Ç–∫–µ [1, 3, 100], –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∫ —Ä–∞–∑—Ä—ã–≤—É —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤ —Ç–æ—á–∫–µ –Ω–æ–ª—å, —Ç.–µ. –ø—Ä–∏ $\sigma\rightarrow ‚àû$ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç.

### –ó–∞–¥–∞—á–∞ 4 (–∫–≤–∞–Ω—Ç–∏–ª—å–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ)

–í –ø—Ä–æ—à–ª–æ–º —Å–µ–º–µ—Å—Ç—Ä–µ –º—ã —Å –≤–∞–º–∏ –≥–æ–≤–æ—Ä–∏–ª–∏ –ø—Ä–æ –∫–≤–∞–Ω—Ç–∏–ª—å–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –Ω–æ –Ω–∏ —Ä–∞–∑—É –µ–≥–æ –Ω–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏. –î–∞–≤–∞–π—Ç–µ —ç—Ç–æ –∏—Å–ø—Ä–∞–≤–∏–º.

–ü—Éc—Ç—å —É –Ω–∞—Å –µ—Å—Ç—å —Å—Ç—Ä—ë–º–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: 

$$
F(x) = \left( \frac{\ln x}{\ln \theta} \right)^{\alpha},  \quad x \in [1; \theta]
$$

__[2.5 –±–∞–ª–ª–∞]__ –°–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –∏–∑ –Ω–µ–≥–æ –≤—ã–±–æ—Ä–∫—É —Å –ø–æ–º–æ—â—å—é –∫–≤–∞–Ω—Ç–∏–ª—å–Ω–æ–≥–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è. –ü–∞—Ä–∞–º–µ—Ç—Ä—ã $\alpha$ –∏ $\theta$ –≤–æ–∑—å–º–∏—Ç–µ –Ω–∞ —Å–≤–æ–π –≤–∫—É—Å.

P.S. [–≤–∏–¥–µ–æ, –∫–æ—Ç–æ—Ä–æ–µ –ø–æ–º–æ–∂–µ—Ç](https://www.youtube.com/watch?v=ivpWyorfWlA&list=PLCf-cQCe1FRyg1ajZ2HJVKknbuTujBOLN&index=11)

```python
uniform_rv = sps.uniform(0, 1)
n_obs = 10**6
y = uniform_rv.rvs(n_obs)
plt.hist(y);
```

```python
theta = 2
alpha = 0.5
x = theta**(y**(1/alpha))
plt.hist(x, density = True, bins = 30);
```

```python
np.mean(x)
```

### –ó–∞–¥–∞—á–∞ 5 (—Å–≤–æ–π—Å—Ç–≤–∞ –æ—Ü–µ–Ω–æ–∫)

–£ –î–∏–∞–Ω—ã –µ—Å—Ç—å –≤—ã–±–æ—Ä–∫–∞ $X_1, \dots, X_n \sim iid N(0, \sigma^2)$. –û–Ω–∞ —Å–æ–±–∏—Ä–∞–µ—Ç—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\sigma$ –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—É—é –æ—Ü–µ–Ω–∫—É $\hat \sigma = \frac{1}{n} \sum_{i=1}^n |x_i|$. 

–ü—É—Å—Ç—å –ø—Ä–∏—Ä–æ–¥–∞ —Ä–µ—à–∏–ª–∞, —á—Ç–æ $\sigma^2 = 16$. 

__–∞) [5 –±–∞–ª–ª–æ–≤]__ –° –ø–æ–º–æ—â—å—é —Å–∏–º—É–ª—è—Ü–∏–π –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –Ω–µ—Å–º–µ—â—ë–Ω–Ω–æ—Å—Ç—å –æ—Ü–µ–Ω–∫–∏, —Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ—Ü–µ–Ω–∫–∏.

__–±) [2.5 –±–∞–ª–ª–∞]__ –†–µ—à–∞—è –∑–∞–¥–∞—á–∫—É —Ä—É–∫–∞–º–∏, –≤—ã –ø–æ–ª—É—á–∏–ª–∏ —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –Ω–∞ —Å–º–µ—â–µ–Ω–∏–µ $\hat \sigma.$ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª—è –Ω–µ—ë –∏ –¥–ª—è –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–π $\hat s$. –ö–∞–∫–∞—è –∏–∑ –Ω–∏—Ö —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–µ–µ? 

P.S. [–Ω–µ—Å–∫–æ–ª—å–∫–æ –≤–∏–¥–µ–æ, –∫–æ—Ç–æ—Ä–æ–µ –ø–æ–º–æ–≥—É—Ç –∏—à–∏ —Ç—É—Ç](https://www.youtube.com/watch?v=arYqiJcyE2s&list=PLCf-cQCe1FRxdjkT1RjdmuFeco11U8ZIP)

```python
sigma = 4
x_data = [sps.norm(loc=0, scale=sigma).rvs(size=i) for i in range(10**3, 10**5, 100)]
sigma_hat =  list(map(lambda x: sum(abs(x)) / len(x), x_data))
```

```python
plt.plot(sigma_hat,'g*', label='sigma_hat')
mean_val = np.mean(sigma_hat)
plt.hlines(y=mean_val, xmin=0, xmax=1000, linestyles='dashed', colors='orange', label='mean value')
plt.ylim([3.1, 3.3])
plt.xlabel('simuls #', fontsize=10)
plt.ylabel('$\hat{\sigma}$', fontsize=10)
plt.legend(loc='lower right')
plt.show()
```

```python
plt.hist(sigma_hat, bins=30);
```

### üêï –î–∞–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –∏–º–µ–µ—Ç —Å–º–µ—â–µ–Ω–Ω–æ—Å—Ç—å, —Ç–∞–∫ –∫–∞–∫ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω—ã—Ö —Å–∏–º—É–ª—è—Ü–∏–π —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–ª–∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –¥–ª—è –Ω–µ–µ ~ 3.192, –∫–æ—Ç–æ—Ä–æ–µ –Ω–µ —Ä–∞–≤–Ω–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—é 4.

```python
import math
sigma_hat_corrected = [np.sqrt(2*math.pi)/2*i for i in sigma_hat]
diffs = [i - sigma for i in sigma_hat_corrected]
```

```python
plt.hlines(y=0, xmin=0, xmax=1000, linestyles='dashed', colors='orange')
plt.ylim([-0.25, 0.25])
plt.xlabel('simuls #', fontsize=10)
plt.ylabel('$\hat{\sigma} - \sigma$', fontsize=10)
plt.plot(diffs);
```

### üêÖ –ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ –≤–∏–¥–Ω–∞ —Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–ª—è $\sigma$.

```python
sigma_hat_classic = []
for i in x_data:
  val = 0
  sz = len(i)
  sm = sum(i)
  for j in range(sz):
     val += (i[j] - 1 / sz * sm)**2
  sigma_hat_classic.append(np.sqrt(1 / (sz - 1) * val))
```

```python
plt.plot(sigma_hat_corrected,'g*', markersize=2.5, label='sigma_hat_corrected')
plt.plot(sigma_hat_classic,'b^', markersize=2.5, label='sigma_hat_classic')
plt.ylim([3.8, 4.2])
plt.legend(loc='lower right')
plt.xlabel('simuls #', fontsize=10)
plt.ylabel('$\hat{\sigma}$', fontsize=10)
plt.show()
```

```python
plt.hist(sigma_hat_corrected,label='sigma_hat_corrected', bins=30)
plt.hist(sigma_hat_classic, label='sigma_hat_classic', bins=30)
plt.legend(loc='lower right')
plt.xlabel('$\hat{\sigma}$', fontsize=10)
plt.ylabel('counts', fontsize=10)
plt.show()
```
